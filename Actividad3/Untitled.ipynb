```python
# Cargar a la memoria de trabajo las bibliotecas: Pandas (manejo de datos, 
# Numpy (numerical python) y la biblioteca de gr√°ficas Matplotlib
# Se asignan nombres cortos.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1= pd.read_csv('junio22.txt', skiprows=5, sep='\s+')
df2= pd.read_csv('diciembre22.txt', skiprows=5, sep='\s+')
df1.columns = ['PRES','HGHT','TEMP','DWPT','RELH','MIXR','DRCT','SKNT','THTA','THTE','THTV']
df2.columns = ['PRES','HGHT','TEMP','DWPT','RELH','MIXR','DRCT','SKNT','THTA','THTE','THTV']
```

```python
df1
```

```python
df2
```

```python
df1 = df1.drop(df1.index[len(df1)-2])
df2 = df2.drop(df2.index[len(df2)-1])
```

```python
df1.dtypes
```

```python
df2.dtypes
```

```python
#Hace que los int64 se conviertan en n√∫mero
columnas = ['HGHT','RELH','DRCT','SKNT']
df1[columnas]= df1[columnas].apply(pd.to_numeric,errors='coerce', axis=1)
df2[columnas]= df2[columnas].apply(pd.to_numeric,errors='coerce', axis=1)
```

```python
#Gr√°fica de la variaci√≥n de la presi√≥n con respecto a la altura para junio
df01 = df1[['HGHT','PRES']]
plt.figure(); df01.plot(x='PRES');plt.legend(loc='best')
plt.title('Variaci√≥n de la Presi√≥n con Respecto a la Altura para junio')
plt.xlim([0,1000])
plt.xlabel('Presi√≥n (hPa)')
plt.ylabel('Altura (m)')
plt.grid(True)
plt.show()
```

```python
#Gr√°fica de la variaci√≥n de la presi√≥n con respecto a la altura para diciembre
df02 = df2[['HGHT','PRES']]
plt.figure(); df02.plot(x='PRES');plt.legend(loc='best')
plt.title('Variaci√≥n de la Presi√≥n con Respecto a la Altura para diciembre')
plt.xlim([0,1000])
plt.xlabel('Presi√≥n (hPa)')
plt.ylabel('Altura (m)')
plt.grid(True)
plt.show()
```

```python
#Gr√°fica de la variaci√≥n de la temperatura con respecto a la altura para junio
df01 = df1[['HGHT','TEMP']]
plt.figure(); df01.plot(x='TEMP');plt.legend(loc='best')
plt.title('Variaci√≥n de la temperatura con respecto a la altura para junio')
plt.xlim([-26.1,30])
plt.xlabel('Temperatura (C)')
plt.ylabel('Altura (m)')
plt.grid(True)
plt.show()
```

```python
#Gr√°fica de la variaci√≥n de la temperatura con respecto a la altura para diciembre
df02 = df2[['HGHT','TEMP']]
plt.figure(); df02.plot(x='TEMP');plt.legend(loc='best')
plt.title('Variaci√≥n de la temperatura con respecto a la altura para diciembre')
plt.xlim([-77,10])
plt.xlabel('Temperatura (C)')
plt.ylabel('Altura (m)')
plt.grid(True)
plt.show()
```

```python
#Gr√°fica de variaci√≥n de la temperatura y de la temperatura de roc√≠o para junio
plt.plot(df1['TEMP'], df1['HGHT'], "red", label='TEMP')
plt.plot(df1['DWPT'], df1['HGHT'], "blue", label='DWPT')
plt.legend(loc='upper right')
plt.title("Gr√°fica Temperatura y Temperatura de roc√≠o con respecto a Altura para junio")
plt.xlabel("Temperatura (¬∞C)")
plt.ylabel("Altura (m)")
plt.grid(True)

plt.show()

```

```python
#Gr√°fica de variaci√≥n de la temperatura y de la temperatura de roc√≠o para diciembre
plt.plot(df2['TEMP'], df2['HGHT'], "red", label='TEMP')
plt.plot(df2['DWPT'], df2['HGHT'], "blue", label='DWPT')
plt.legend(loc='upper right')
plt.title("Gr√°fica Temperatura y Temperatura de roc√≠o con respecto a Altura para diciembre")
plt.xlabel("Temperatura (¬∞C)")
plt.ylabel("Altura (m)")
plt.grid(True)

plt.show()

```

```python
#Gr√°fica de la variaci√≥n de la rapidez de los vientos con respecto a la altura para junio
df01 = df1[['HGHT','SKNT']]
plt.figure(); df01.plot(x='SKNT');plt.legend(loc='best')
plt.title('Variaci√≥n de la rapidez de los vientos con respecto a la altura para junio')
plt.xlim([0,50])
plt.xlabel('Rapidez de los vientos (knot)')
plt.ylabel('Altura (m)')
plt.grid(True)
plt.show()
```

```python
#Gr√°fica de la variaci√≥n de la rapidez de los vientos con respecto a la altura para diciembre
df02 = df2[['HGHT','SKNT']]
plt.figure(); df02.plot(x='SKNT');plt.legend(loc='best')
plt.title('Variaci√≥n de la rapidez de los vientos con respecto a la altura para diciembre')
plt.xlim([0,67])
plt.xlabel('Rapidez de los vientos (knot)')
plt.ylabel('Altura (m)')
plt.grid(True)
plt.show()
```

```python
#Gr√°fica de la variaci√≥n de la humedad relativa con respecto a la altura para junio
df01 = df1[['HGHT','RELH']]
plt.figure(); df01.plot(x='RELH');plt.legend(loc='best')
plt.title('Variaci√≥n de la humedad relativa con respecto a la altura para junio')
plt.xlim([0,70])
plt.xlabel('Humedad relativa (%)')
plt.ylabel('Altura (m)')
plt.grid(True)
plt.show()
```

```python
#Gr√°fica de la variaci√≥n de la hmedad relativa con respecto a la altura para diciembre
df02 = df2[['HGHT','RELH']]
plt.figure(); df02.plot(x='RELH');plt.legend(loc='best')
plt.title('Variaci√≥n de la humedad relativa con respecto a la altura para diciembre')
plt.xlim([0,95])
plt.xlabel('Humedad relativa (%)')
plt.ylabel('Altura (m)')
plt.grid(True)
plt.show()
```
