```python
import pandas as pd
import numpy as np
from datetime import datetime
import seaborn as sns
```

```python
df1= pd.read_csv('sargento-270218.csv', header=None, skiprows=2, sep=',')
df2= pd.read_csv('sargento-salinidad-270218.csv', header=None, skiprows=2, sep=',')
```

```python
df1.columns = ['NO','Date','abs','temp','Met']
df2.columns = ['NO','Date','CondH','temp','Conduc','sal']
```

```python
df1.head()
```

```python
# Convertir la cadena de caracteres 'Date' en variable temporal 'NDate'
df1['Ndate'] = pd.to_datetime(df1['Date'], format='%m/%d/%Y %H:%M:%S')
df1['month'] = df1['Ndate'].dt.month
df1.head()
```

```python
# Convertir la cadena de caracteres 'Date' en variable temporal 'NDate'
df2['Ndate'] = pd.to_datetime(df2['Date'], format='%m/%d/%Y %H:%M:%S')
df2['month'] = df2['Ndate'].dt.month
df2.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
ax = sns.boxplot(x="month", y="Met", data=df1)
plt.show()
```
